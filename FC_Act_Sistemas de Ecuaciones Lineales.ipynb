**Paso 1:** Encontrar y reportar la solución a los sistemas (solo usar hoja y pluma)


$(i)$
    \begin{align}
        4x+2y=4\\
        2x+y=1
    \end{align}
Multiplicamos la primera ecuación por $\displaystyle\frac{1}{2}$ y obtenemos $2x+y=2$. Si a esta expresión le restamos nuestra segunda ecuación obtenemos que $0=1$, lo que nos indica que no existe solución para este sistema.

$(ii)$
    \begin{align}
        x+2y+2z=2\\
        2x+4y+4z=4\\
        4x+2y+z=0
    \end{align}
Multiplicamos la primera ecuación por $\displaystyle\frac{1}{2}$ y obtenemos $x+2y+2z=2$, que es nuestra primera ecuación. Por lo que la primera y segunda ecuación son reduntantes, por lo que nuestro sistema es en realidad de dos ecuaciones, y como tenemos tres variables, podemos anticipar que la solución no será única.

Buscamos entonces una solución para el siguiente sistema:
\begin{align}
    x+2y+2z=2\\
    4x+2y+z=0
\end{align}
Restamos la segunda de la primera y nos quedamos con
\begin{align}
    -3x+z=2
\end{align}
Despejando para $z$ obtenemos
\begin{align}
    z=3x+2
\end{align}
Podemos sustituir esto en la segunda ecuación para despejar para $y$
\begin{align}
    4x+2y+(3x+2)=0\\
    \Rightarrow y=-\frac{7}{2}x-1
\end{align}
Para estar seguros sustituimos también en la primera ecuación
\begin{align}
    x+2y+2(3x+2)=2\\
    \Rightarrow 2y=2-7x-4\\
    \Rightarrow y=\frac{7}{2}x-2
\end{align}
Con lo que confirmamos y podemos escribir la solución como un vector columna de la siguiente manera
\begin{align}
\begin{bmatrix}
    x\\
    -3.5x-4\\
    3x+2
\end{bmatrix}\qquad \forall x\in\mathbb{R}
\end{align}

Ahora la tercia de planos

\begin{align}
        x+2y+2z=2\\
        2x+4y+4z=4\\
        4x+2y+z=0
\end{align}

Debemos despejar para $z$ (en realidad podría ser para cualquier otra variable, pero por convención escogemos esta) y nos queda

\begin{align}
        z=1-\frac{1}{2}x-y\\
        z=1-\frac{1}{2}x-y\\
        z=-4x-2y
\end{align}

```python
from mpl_toolkits import mplot3d
def f1(x, y):
    return 1-x/2-y
def f2(x, y):
    return 1-x/2-y
def f3(x, y):
    return -4*x-2*y
x= np.linspace(-10, 10, 100)
y= np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z1 = f1(X, Y)
Z2 = f2(X, Y)
Z3 = f3(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z1, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.plot_surface(X, Y, Z2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.plot_surface(X, Y, Z3, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

# Actividad 2
Resolver mediante Eliminación Gaussiana el sistema:
\begin{align}
    0.0003x_1+3.0000x_2=2.0001\\
    1.0000x_1+1.0000x_2=1.0000
\end{align}

Primero que nada, por facilidad, lo escribimos en formato de matriz y empezamos con la primera columna como pivote
\begin{align}
\begin{bmatrix}
    0.0003&3.0000&2.0001\\
    1.0000&1.0000&1.0000
\end{bmatrix}\qquad
R_2-\frac{1.0000}{0.0003}\cdot R_1\\
\Rightarrow
\begin{bmatrix}
    0.0003&3.0000&2.0001\\
    0&-9,666&-6,666
\end{bmatrix}\\
\Rightarrow x_2=\frac{-6,666}{-9,999}=\frac{2}{3}=0.6667\\
\Rightarrow 0.0003x_1+3.0000(0.6667)=2.0001\\
\Rightarrow 0.0003x_1=0\\
\Rightarrow x_1=0
\end{align}

Ahora utilizamos la segunda columna como pivote

\begin{align}
\begin{bmatrix}
    1.0000&1.0000&1.0000\\
    0.0003&3.0000&2.0001
\end{bmatrix}\qquad
R_2-\frac{0.0003}{1.0000}\cdot R_1\\
\Rightarrow\begin{bmatrix}
    1.0000&1.0000&1.0000\\
    0&2.9997&1.9998
\end{bmatrix}\\
\Rightarrow x_2=\frac{1.9998}{2.9997}=\frac{2}{3}=0.6667\\
\Rightarrow x_1+0.6667=1\\
\Rightarrow x_1=0.3333
\end{align}

Notamos que los resultados son diferentes, más aún que el primero no satisface la ecuación, mientras que el segundo sí.

### Reto en Python

Resolverlo con $\displaystyle x_2=\frac{2}{3}$ considerando diferente número de cifras significativas.

```python
#Al ejecutar la celda se solicita al usuario el número n para el número de cifras
A = np.array([[0.0003, 3.0000, 2.0001],[1.0000, 1.0000, 1.0000]]);
n=int(input("Número de cifras significativas: "))
#Resta de la segunda columna con un escalamiento de la primera
r1 = (A[1,0]/A[0,0])*A[0];
for j in range (0,1):
        r1[j]=round(r1[j],n)
R1 = np.array([[0,0,0],r1]);
A = A - R1;
for i in range (0,2):
    for j in range (0,1):
        A[i,j]=round(A[i,j],n)
print("Tras aplicar Eliminación Gaussiana=",A);
x2=A[1,2]/A[1,1]
x2=round(x2,n)
print("x2 =",x2)
print("Redondeando a n =",n,"obtenemos que: ")
x1=(A[0,2]-(A[0,1]*x2))/A[0,0]
x1=round(x1,n)
print("x1 =",x1)
```
